# Binary classification based on Logistic Regression with a quadratic regularization

## import library

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
```

## load data

```python
fname_data_train    = 'assignment_10_data_train.csv'
fname_data_test     = 'assignment_10_data_test.csv'

data_train  = np.genfromtxt(fname_data_train, delimiter=',')
data_test   = np.genfromtxt(fname_data_test, delimiter=',')
```

## plot the data

```python
# data 분류
# data1 = train, data2 = test

num_data1, num_data2 = data_train.shape[0], data_test.shape[0]

data1_zero, data1_one = [],[]
data2_zero, data2_one = [],[]

for data in data_train:
    if data[2] == 0:
        data1_zero.append(data[:2])
    else:
        data1_one.append(data[:2])
for data in data_test:
    if data[2] == 0:
        data2_zero.append(data[:2])
    else:
        data2_one.append(data[:2])
        
data1_zero, data1_one = np.array(data1_zero), np.array(data1_one)
data2_zero, data2_one = np.array(data2_zero), np.array(data2_one)

# plot
plt.figure(figsize=(16,8))

# data1
plt.subplot(121)
plt.plot(data1_zero[:,0], data1_zero[:,1], '.', color = 'blue', label='class 0')
plt.plot(data1_one[:,0], data1_one[:,1], '.', color = 'red', label='class 1')
plt.title('training data')
plt.legend()

# data2
plt.subplot(122)
plt.plot(data2_zero[:,0], data2_zero[:,1], '.', color = 'blue', label='class 0')
plt.plot(data2_one[:,0], data2_one[:,1], '.', color = 'red', label='class 1')
plt.title('testing data')

plt.legend()
#plt.axis('equal')
plt.tight_layout()
plt.show()

```

## define feature function

```python
def feature_function(x, y):
    
    feature = np.array([[1]*num_data2, x, x**2, x**3, y])
    
    return feature
```

## define regression function based on the feature function

```python
def regression_function(theta, feature):
    
    value = feature.T @ theta
    
    return value
```

## define regularization function on the model parameters

```python
def regularization_function(theta):
    #
    value = np.sum(theta**2)
    #
    return value
```

## define sigmoid function

```python
def logistic_function(x):
    
    z = 1 / (1+np.exp(-x))
    
    return z
```

## define loss function where $\alpha$ is a weight for the quadratic regularization term (Note that you need to add a small number (np.finfo(float).eps) inside logarithm function in order to avoid $\log( 0 )$)

```python
def compute_loss_feature(theta, feature, label, alpha):
    
    wTx = regression_function(theta, feature)
    h = logistic_function(wTx)
    r = regularization_function(theta)
    num_data = feature.shape[1]
    loss = -(1/num_data)*np.sum(label*np.log(h)+(np.full(num_data,1)-label)*np.log(np.full(num_data,1)-h)) + (alpha/2)*r
    
    return loss
```

## define gradient vector for the model parameters with the quadratic regularization term whose weight is $\alpha$

```python
def compute_gradient_feature(theta, feature, label, alpha):
    
    wTx = regression_function(theta, feature)
    h = logistic_function(wTx)
    num_data = feature.shape[1]
    gradient = (1/num_data) * feature@(h-label) + alpha*theta
    
    return gradient
```

## compute the accuracy

```python
def compute_accuracy(theta, feature, label):
    
    tot = len(label)
    r = regression_function(theta, feature)
    pred = []
    e = 0
    for val in r:
        if val<=0:
            pred.append(0)
        else:
            pred.append(1)
    for i in range(tot):
        if label[i] != pred[i]:
            e +=1
    accuracy = 1 - (e/tot)
    
    return accuracy
```

## gradient descent for the model parameters $\theta$

```python
num_iteration   = 19000
learning_rate   = 0.1
alpha           = 0
#
theta = np.array((0, 0, 0, 0, 0))
feature_train = feature_function(data_train[:,0], data_train[:,1])
feature_test = feature_function(data_test[:,0], data_test[:,1])
#
```

```python
theta_iteration             = np.zeros((num_iteration, theta.size))
loss_iteration_train        = np.zeros(num_iteration)
loss_iteration_test         = np.zeros(num_iteration)
accuracy_iteration_train    = np.zeros(num_iteration)
accuracy_iteration_test     = np.zeros(num_iteration)
```

```python
np.seterr(divide = 'ignore')
np.seterr(invalid='ignore')

for i in range(num_iteration):
    
    theta = theta - learning_rate * compute_gradient_feature(theta, feature_train, data_test[:,2], alpha)
    
    loss_train = compute_loss_feature(theta, feature_train, data_train[:,2], alpha)
    loss_test = compute_loss_feature(theta, feature_test, data_test[:,2], alpha)
    
    accuracy_train = compute_accuracy(theta, feature_train, data_train[:,2])
    accuracy_test = compute_accuracy(theta, feature_test, data_test[:,2])
    
    
    theta_iteration[i] = theta
    loss_iteration_train[i] = loss_train
    loss_iteration_test[i] = loss_test
    accuracy_iteration_train[i] = accuracy_train
    accuracy_iteration_test[i] = accuracy_test
    theta_optimal = theta
    
    #print("loss(train) = %5.5f, loss(test) = %5.5f, accuracy(train) = %5.5f, accuracy(test) = %5.5f" % (loss_train, loss_test, accuracy_train, accuracy_test))
```

## compute accuracy of the classifiers

```python
accuracy_train  = compute_accuracy(theta_optimal, feature_train, data_train[:,2])
accuracy_test   = compute_accuracy(theta_optimal, feature_test, data_test[:,2])
```
